 The filtered dataset except mobile_number and churn columns all the columns are numeric types. Hence, converting mobile_number and churn datatype to object.

```python
data_telecom['mobile_number'] = data_telecom['mobile_number'].astype(object)
data_telecom['churn'] = data_telecom['churn'].astype(object)
```

```python
# Removing outliers below 10th and above 90th percentile
for col in numeric_cols: 
    q1 = data_telecom[col].quantile(0.10)
    q3 = data_telecom[col].quantile(0.90)
    iqr = q3-q1
    range_low  = q1-1.5*iqr
    range_high = q3+1.5*iqr
    # Assigning the filtered dataset into data
    data_df = data_telecom.loc[(data_telecom[col] > range_low) & (data_telecom[col] < range_high)]

data_df.shape
```

# Implementing new column decrease_mou_action

This column indicates whether the minutes of usage of the customer has decreased in the action phase than the good phase.

```python
# Total mou at good phase incoming and outgoing
data_df['total_mou_good'] = (data_df['total_og_mou_6'] + data_df['total_ic_mou_6'])
```

```python
# Avg. mou at action phase
# now here taking average because there are two months(7 and 8) in action phase
data_df['avg_mou_action'] = (data_df['total_og_mou_7'] + data_df['total_og_mou_8'] + data_df['total_ic_mou_7'] + data_df['total_ic_mou_8'])/2
```

```python
# Difference avg_mou_good and avg_mou_action.
data_df['diff_mou'] = data_df['avg_mou_action'] - data_df['total_mou_good']
```

```python
# now we are checking for the mou has decreased in action phase.
data_df['decrease_mou_action'] = np.where((data_df['diff_mou'] < 0), 1, 0)
```

```python

data_df.head()
```

# Deriving new column decrease_rech_num_action

This column indicates whether the number of recharge of the customer has decreased in the action phase than the good phase.

```python
# Avg rech number at action phase
data_df['avg_rech_num_action'] = (data_df['total_rech_num_7'] + data_df['total_rech_num_8'])/2
```

```python
# Difference total_rech_num_6 and avg_rech_action
data_df['diff_rech_num'] = data_df['avg_rech_num_action'] - data_df['total_rech_num_6']
```

```python
# Checking if rech_num has decreased in action phase
data_df['decrease_rech_num_action'] = np.where((data_df['diff_rech_num'] < 0), 1, 0)
```

```python
data_df.head()
```

# Deriving new column decrease_rech_amt_action

This column indicates whether the amount of recharge of the customer has decreased in the action phase than the good phase.

```python
#  rech_amt in action phase in avg
data_df['avg_rech_amt_action'] = (data_df['total_rech_amt_7'] + data_df['total_rech_amt_8'])/2
```

```python
# Difference of action phase rech amt and good phase rech amt
data_df['diff_rech_amt'] = data_df['avg_rech_amt_action'] - data_df['total_rech_amt_6']
```

```python
#  if rech_amt has decreased in action phase
data_df['decrease_rech_amt_action'] = np.where((data_df['diff_rech_amt'] < 0), 1, 0) 
```

```python
data_df.head()
```

# Deriving new column decrease_arpu_action

The column indicates whether the average revenue per customer has decreased in the action phase than the good phase.

```python
#  action phase ARUP 
data_df['avg_arpu_action'] = (data_df['arpu_7'] + data_df['arpu_8'])/2
```

```python
#  good and action phase ARPU
data_df['diff_arpu'] = data_df['avg_arpu_action'] - data_df['arpu_6']
```

```python
# Checking whether the arpu has decreased on the action month
data_df['decrease_arpu_action'] = np.where(data_df['diff_arpu'] < 0, 1, 0)
```

```python
data_df.head()
```

# New column decrease_vbc_action

The column indicates whether the volume based cost of the customer has decreased in the action phase than the good phase.

```python
# VBC in action phase
data_df['avg_vbc_3g_action'] = (data_df['jul_vbc_3g'] + data_df['aug_vbc_3g'])/2
```

```python
#  good and action phase VBC
data_df['diff_vbc'] = data_df['avg_vbc_3g_action'] - data_df['jun_vbc_3g']
```

```python
#  whether the VBC has decreased on the action month lets check
data_df['decrease_vbc_action'] = np.where(data_df['diff_vbc'] < 0 , 1, 0)
```

Optimal Probablity Cutoff Point

```python
# Creating columns for different probablity cutoffs
prob_cutoff = [float(p/10) for p in range(10)]

for i in prob_cutoff:
    y_train_pred_final[i] = y_train_pred_final['churn_prob'].map(lambda x : 1 if x > i else 0)
    
y_train_pred_final.head()
```

```python
# Creating a dataframe
cutoff_df = pd.DataFrame(columns=['probability', 'accuracy', 'sensitivity', 'specificity'])

for i in prob_cutoff:
    cm1 = metrics.confusion_matrix(y_train_pred_final['churn'], y_train_pred_final[i] )
    total1=sum(sum(cm1))
    accuracy = (cm1[0,0]+cm1[1,1])/total1
    
    speci = cm1[0,0]/(cm1[0,0]+cm1[0,1])
    sensi = cm1[1,1]/(cm1[1,0]+cm1[1,1])
    cutoff_df.loc[i] =[ i ,accuracy,sensi,speci]
print(cutoff_df)
    
```

```python
# Plotting accuracy, sensitivity and specificity for different probabilities.
cutoff_df.plot('probability', ['accuracy','sensitivity','specificity'])
plt.show()
```

Analysis of the above curve
Accuracy - Becomes stable around 0.6

Sensitivity - Decreases with the increased probablity.

Specificity - Increases with the increasing probablity.

At point 0.6 where the three parameters cut each other, we can see that there is a balance bethween sensitivity and specificity with a good accuracy.

Here we are intended to acheive better sensitivity than accuracy and specificity. Though as per the above curve, we should take 0.6 as the optimum probability cutoff, we are taking *0.5* for acheiving higher sensitivity, which is our main goal.

# ROC Curve (Trade off between sensitivity & specificity)

```python
# ROC Curve function

def draw_roc( actual, probs ):
    fpr, tpr, thresholds = metrics.roc_curve( actual, probs,
                                              drop_intermediate = False )
    auc_score = metrics.roc_auc_score( actual, probs )
    plt.figure(figsize=(5, 5))
    plt.plot( fpr, tpr, label='ROC curve (area = %0.2f)' % auc_score )
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characteristic example')
    plt.legend(loc="lower right")
    plt.show()

    return None
```
